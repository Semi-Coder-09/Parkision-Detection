# IMPORT THE DEPENDENCIES

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
from sklearn.pipeline import make_pipeline
```

# DATA COLLECTION AND ANALYSIS

```python
df = pd.read_csv('merged_dataset.csv')

# Basic statistics of the dataset
print(df.describe())
```

```python
# number of rows and columns in the dataframe
df.shape
```

```python
# getting more information about the dataset
df.info()
```

```python
# checking for missing values in each column
df.isnull().sum()
```

```python
# getting some statistical measures about the data
df.describe()
```

```python
# distribution of target Variable
df['status'].value_counts()
```

```python
df['status'] = pd.to_numeric(df['status'], errors='coerce')

# Now you can apply the mean function
mean_value = df['status'].mean()
```

```python
X = df.drop(columns=['name','status'], axis=1)
Y = df['status']
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import StandardScaler

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Standardizing the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print(X_train)
```

# MODEL TRAINING (DECISION TREE)

```python
DT_pipe = make_pipeline(DecisionTreeClassifier(criterion='entropy'))
DT_pipe.fit(X_train, y_train)
y_pred = DT_pipe.predict(X_test)
```

```python
# Performance Measures
score = DT_pipe.score(X_test, y_test)
acc_score = accuracy_score(y_test, y_pred)
PRF_Micro = precision_recall_fscore_support(y_test, y_pred, average='micro')
PRF_Macro = precision_recall_fscore_support(y_test, y_pred, average='macro')
```

```python
# Cross-Validation
cv_scores = cross_val_score(DT_pipe, X, Y, cv=10)

# Print performance metrics
print("Accuracy Score:", score)
print("Accuracy:", acc_score)
print("Precision Recall F-Measure:")
print("Micro:", PRF_Micro)
print("Macro:", PRF_Macro)
print("CV-Score:", cv_scores)
print('CV Scores Mean:', np.mean(cv_scores))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")
# Input data for prediction
input_data = input("Enter the data for prediction (comma-separated values): ")
input_list = input_data.split(",")
input_data_as_numpy_array = np.array(input_list)

# Reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Standardize the data
std_data = scaler.transform(input_data_reshaped)

# Make prediction
prediction = DT_pipe.predict(std_data)

# Print prediction result
if prediction[0] == 0:
    print("The Person does not have Parkinson's Disease")
else:
    print("The Person has Parkinson's")
```

```python
# Plot confusion matrix for Decision Tree
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=[0, 1], yticklabels=[0, 1])
plt.title("Confusion Matrix (Decision Tree)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
