# IMPORT THE DEPENDENCIES

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import make_pipeline
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")
warnings.filterwarnings("ignore", category=FutureWarning)
```

# DATA COLLECTION AND ANALYSIS

```python
df = pd.read_csv('merged_dataset.csv')

# Basic statistics of the dataset
print(df.describe())
```

```python
df['status'] = pd.to_numeric(df['status'], errors='coerce')

# Now you can apply the mean function
mean_value = df['status'].mean()
```

```python
X = df.drop(columns=['name','status'], axis=1)
Y = df['status']
print(X)
```

```python
print(Y)
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Standardizing the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print(X_train)
```

# MODEL TRAINING (BAGGING)

```python
svc_base = SVC(probability=True, kernel='linear')
rf_base = RandomForestClassifier()

# Creating bagging classifiers
bagging_svc = BaggingClassifier(base_estimator=svc_base, n_estimators=10, random_state=42)
bagging_rf = BaggingClassifier(base_estimator=rf_base, n_estimators=10, random_state=42)

# Creating pipelines without standard scaling (data is already scaled during bagging)
svc_pipe = make_pipeline(bagging_svc)
rf_pipe = make_pipeline(bagging_rf)

# Fitting and predictiction of SVM-based Bagging
svc_pipe.fit(X_train, y_train)
y_pred_svc = svc_pipe.predict(X_test)

# Fitting and prediction of RF-based Bagging
rf_pipe.fit(X_train, y_train)
y_pred_rf = rf_pipe.predict(X_test)

# Performance Measures for SVM-based Bagging
score_svc = svc_pipe.score(X_test, y_test)
acc_score_svc = accuracy_score(y_test, y_pred_svc)
PRF_Micro_svc = precision_recall_fscore_support(y_test, y_pred_svc, average='micro')
PRF_Macro_svc = precision_recall_fscore_support(y_test, y_pred_svc, average='macro')
cv_scores_svc = cross_val_score(svc_pipe, X, Y, cv=10)

# Performance Measures for RF-based Bagging
score_rf = rf_pipe.score(X_test, y_test)
acc_score_rf = accuracy_score(y_test, y_pred_rf)
PRF_Micro_rf = precision_recall_fscore_support(y_test, y_pred_rf, average='micro')
PRF_Macro_rf = precision_recall_fscore_support(y_test, y_pred_rf, average='macro')
cv_scores_rf = cross_val_score(rf_pipe, X, Y, cv=10)
```

```python
ensemble_bagging = BaggingClassifier(base_estimator=svc_pipe, n_estimators=10, random_state=42, n_jobs=-1)
ensemble_bagging.fit(X_train, y_train)

# Predictions
y_pred_ensemble = ensemble_bagging.predict(X_test)

# Performance Measures for Ensemble Bagging
score_ensemble = ensemble_bagging.score(X_test, y_test)
acc_score_ensemble = accuracy_score(y_test, y_pred_ensemble)
PRF_Micro_ensemble = precision_recall_fscore_support(y_test, y_pred_ensemble, average='micro')
PRF_Macro_ensemble = precision_recall_fscore_support(y_test, y_pred_ensemble, average='macro')
cv_scores_ensemble = cross_val_score(ensemble_bagging, X, Y, cv=10)

# Ensemble Bagging results
print("\nEnsemble Bagging (SVM and RF):")
print("Accuracy Score:", score_ensemble)
print("Accuracy:", acc_score_ensemble)
print("Precision Recall F-Measure:")
print("Micro:", PRF_Micro_ensemble)
print("Macro:", PRF_Macro_ensemble)
print("CV-Score:", cv_scores_ensemble)
print('CV-Scores mean:', np.mean(cv_scores_ensemble))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Create a confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_ensemble)

# Plot the confusion matrix using seaborn
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['0', '1'], yticklabels=['0', '1'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
